```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7] # day
y=[50,51,52,48,47,49,46] # temperature
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, 'go-.')
```

```python
# equivalent effect using plot api
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='blue', marker='+', linestyle='', markersize=30)
#plt.plot(x,y, 'b+')
```

```python
plt.plot(x,y, 'g<', alpha=0.5)
```
