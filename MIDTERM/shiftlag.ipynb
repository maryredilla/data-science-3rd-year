```python
import pandas as pd
df = pd.read_csv("fb_shiftlag.csv", parse_dates=['Date'], index_col='Date')
df
```

```python
df.shift(1)
```

```python
df.shift(-1)
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['Price Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
# (price current - prev(5 days) * 10)/ prev price(5 days)

df['5 day return'] = (df['Price'] - df['Price'].shift(5))*100
df
```
