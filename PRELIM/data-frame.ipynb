## **Creating Data Frame**

```python
# using csv
import pandas as pd
df = pd.read_csv('sample_weather_data01.csv')
df
```

```python
# using excel
df = pd.read_excel("sample_weather_data.xlsx", "Sheet1")
df
```

```python
# manual input/ using dictionaries
sample_weather_data ={
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain', 'Sunny', 'Snow']
}
df = pd.DataFrame(sample_weather_data)
df
```

```python
# using tuple list
sample_weather_data = {
    ('1/1/2017', 32,6,'Rain'),
    ('1/2/2017', 35,7,'Sunny'),
    ('1/3/2017', 28,2,'Snow')
}

df = pd.DataFrame(data=sample_weather_data,columns=['day', 'temperature', 'windspeed', 'event'])
df
```

```python
# using list of dictionaries
# keys(day) - values
sample_weather_data = [
    {'day': '1/1/2017', 'temperature': '32', 'windspeed': '6', 'event': 'Rain'}
]

df = pd.DataFrame(data=sample_weather_data, columns=['day', 'temperature', 'windspeed', 'event'])
df
```
