```python
import pandas as pd

df1 = pd.DataFrame({
    "city": ["Cebu", "Manila", "QC"],
    "temperature": [29, 35, 31],
})

df1
```

```python
df2 = pd.DataFrame({
    "city": ["Cebu", "Manila", "QC"],
    "temperature": [55, 65, 66],
})

df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
df2 = pd.DataFrame({
    "city": ["Cebu", "Manila", "QC", "Pasig"],
    "humidity": [65, 70, 73, 33],
})

df2
```

```python
df3 = pd.merge(df1, df2, on="city", how='inner')
df3
```

```python
df3 = pd.merge(df1, df2, on="city", how='outer')
df3
```

```python
df3 = pd.merge(df1, df2, on="city", how='right')
df3
```

```python
df3 = pd.merge(df1, df2, on="city", how='outer', indicator=True)
df3
```

```python
df3 = pd.DataFrame({
    "city": ["Manila", "Cebu", "Tanauan"],
    "temperature": [29, 35, 31],
    "humidity": [65, 70, 73],
})

df3
```

```python
df3 = pd.merge(df1, df2, on="city", how='outer', suffixes=('_first', '_second'))
df3
```
