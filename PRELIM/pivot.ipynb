## Pivot

```python
import pandas as pd

import pandas as pd
import numpy as np

df = pd.read_csv('pivot_weather01.csv')
df
```

```python
df.pivot(index='date', columns='city')
```

```python
# Humidity
df.pivot(index='city', columns='date', values='humidity')
```

```python
# Temperature
df.pivot(index='city', columns='date', values='temperature')
```

```python
df_2 = pd.read_csv('pivot_weather02.csv')
df_2
```

```python
df_2.pivot_table(index='city', columns='date')
```

## Margins

```python
# margins = mean
df_2.pivot_table(index='city', columns='date', margins=True)
```

## Aggregation Function

```python
# aggregation function (numpy) - mean
df_2.pivot_table(index='city', columns='date', margins=True, aggfunc=np.sum)

# aggfunc=np.std
# aggfunc=np.average
```
