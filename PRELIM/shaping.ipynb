## Crosstabulation

```python
# crosstab to check if data sets are balanced
import pandas as pd

df_4 = pd.read_excel('crosstab_survey.xlsx')
df_4
```

```python
pd.crosstab(df_4.Nationality, df_4.Handedness)
```

```python
pd.crosstab(df_4.Sex, df_4.Handedness)
```

```python
pd.crosstab(df_4.Sex, df_4.Handedness, margins=True)
```

## Multi-index Columns and Rows

```python
pd.crosstab([df_4.Nationality, df_4.Sex], df_4.Handedness, margins=True)
```

## Normalization

```python
pd.crosstab(df_4.Sex, df_4.Handedness, normalize='index')
```

## Aggregation

```python
import numpy as np
pd.crosstab(df_4.Sex, df_4.Handedness, values=df_4.Age, aggfunc=np.average)
```
