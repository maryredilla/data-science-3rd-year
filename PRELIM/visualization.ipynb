```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7] # day
y = [50,51,52,53,47,49,46] # temperature
```

```python
# plot x and y

plt.plot(x,y, color='green', linewidth=3, linestyle='dotted')
```

```python
plt.plot(x,y, color='green', linewidth=3, linestyle=':')
```

```python
plt.plot(x,y, color='green', linewidth=3, linestyle='dashed')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, color='purple', linewidth=3, linestyle='solid')
```
