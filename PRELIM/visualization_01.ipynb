```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7] # day
y = [50,51,52,53,47,49,46] # temperature
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, 'b') # blue, solid line
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, 'b--') # blue, dotted line
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, 'b+--') # blue, dotted line, plus marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y, 'ro--') # red, dotted line, circle marker
```

```python
# equivalent effect plot api

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='blue', marker='+', linestyle='')
plt.plot(x,y, 'b+') # blue, + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='blue', marker='+', linestyle='', markersize=10)
```

```python
plt.plot(x,y, 'g<', alpha=0.2) # transparent
```
