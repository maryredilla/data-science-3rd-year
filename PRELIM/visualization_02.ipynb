```python
# histogram for statistics

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8, bins=5)
```

```python
# 80 - 100 (Normal)
# 101 -125 (Pre-diabetic)
# 126 - 150 (Diabetic)

plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.90, color='pink')
```

```python
# 80 - 100 (Normal)
# 101 -125 (Pre-diabetic)
# 126 - 150 (Diabetic)

plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist(blood_sugar_men, blood_sugar_women, bins=[80,100,125,150], rwidth=0.90, color=['blue','pink'])
plt.legend(shadow='True')
```
