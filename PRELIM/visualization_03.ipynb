```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
```

```python
import numpy as np
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos, revenue, label="Revenue")
plt.xticks(xpos, company)
```

```python
plt.bar(xpos, revenue, label="Revenue")
plt.xticks(xpos, company)
plt.xlabel('Revenue')
plt.ylabel('Tech Company')

plt.legend()
```

```python
profit=[40,3,34,12]
```

```python
plt.bar(xpos-0.2, revenue, width=0.2, label="Revenue")
plt.bar(xpos+0.2, profit, width=0.2, label="Profit")

plt.xticks(xpos, company)
plt.xlabel('Revenue')
plt.ylabel('Tech Company')

plt.legend()
```
