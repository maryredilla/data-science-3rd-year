### Labels and Legends

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7] # day
max_t = [50,51,52,53,47,49,46] # temperature
min_t = [43, 42, 40, 44, 33, 35,37]
avg_t = [45, 48, 48, 46, 40, 42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend(loc='best', shadow='True', fontsize='x-small')
```
